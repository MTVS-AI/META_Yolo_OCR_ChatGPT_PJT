```python
origin_path = 'data_002.jpg'
detect_path = 'data_0020.jpg'
crop_frame = 'data_0024.jpg'
crop_path_1 = 'data_0021.jpg'
crop_path_2 = 'data_0022.jpg'
crop_path_3 = 'data_0023.jpg'
```

```python
df.iloc[1]
```

```python
from PIL import Image
import numpy as np

df['Origin_img'][1] = [np.array(Image.open(origin_path)).tolist()]
df['Detect_img'][1] = [np.array(Image.open(detect_path)).tolist()]
df['Crop_imgs'][1] = [[np.array(Image.open(crop_frame)).tolist()], [np.array(Image.open(crop_path_1)).tolist()], [np.array(Image.open(crop_path_2)).tolist()], [np.array(Image.open(crop_path_3)).tolist()]]
df.iloc[1]
```

```python
df['Location'][1] = [37.394775, 127.11126]
```

```python
df.iloc[1]['Location']
```

```python
img_path = [origin_path, detect_path, crop_frame, crop_path_1, crop_path_2, crop_path_3]
```

```python
origin_path[:-3]
```

```python
images = []
for img in img_path:
    image = Image.open(img)
    image = image.resize((int(image.size[0]/2), int(image.size[1]/2)))
    image.save(img, 'jpeg')
    images.append(image)
```

```python
new_img_path = []
for img in img_path:
    image = Image.open(img).convert('RGB')
    image = image.resize((int(image.size[0]/2), int(image.size[1]/2)))
    new_img = img[:-3] + 'png'
    image.save(new_img, 'png')
    new_img_path.append(new_img)
```

```python
import base64

pic_list = []
for img in img_path:
    pic = base64.b64encode(open(img, 'rb').read()).decode()
    pic_list.append(pic)
```

```python
i = Image.open(new_img_path[0])
i.size[1]
```

```python
import folium

center = [37.394776, 127.11116]
map = folium.Map(location = center, zoom_start=12)

image_tag = '<div style="display: flex; justify-content: flex-end;"><img src="data:image/jpeg;base64,{0}"><img src="data:image/jpeg;base64,{1}"></div><div style="display: flex; justify-content: flex-end;"><img src="data:image/jpeg;base64,{2}"><img src="data:image/jpeg;base64,{3}"><img src="data:image/jpeg;base64,{4}"><img src="data:image/jpeg;base64,{5}"></div>'.format(pic_list[0], pic_list[1], pic_list[2], pic_list[3], pic_list[4], pic_list[5])
iframe = folium.IFrame(image_tag, width=1000, height=500)
popup = folium.Popup(iframe, max_width=1000)

folium.Marker(location = df.iloc[1]['Location'], icon = folium.Icon(color = 'blue', icon = 'star'), popup = popup).add_to(map)

map
```
